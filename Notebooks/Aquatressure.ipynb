The dataset is dominated by Perch and Bream species, indicating these fish may be more commonly sampled or commercially significant.


## 2. What Is the Average Weight for Each Species?

```python
df.groupby("Species")["Weight"].mean().sort_values(ascending=False)
```

```python
sns.barplot(data=df, x="Species", y="Weight")
plt.xticks(rotation=45)
plt.title("Average Weight by Fish Species")
plt.show()
```

Certain species such as Pike tend to have higher average weights compared to smaller species like Smelt.

## 3. How Is Fish Weight Distributed?

```python
sns.histplot(df["Weight"], kde=True)
plt.title("Fish Weight Distribution")
plt.show()
```

The weight distribution appears right-skewed, meaning a small number of fish have significantly larger weights.

## 4. Do Larger Fish Always Weigh More?

```python
sns.scatterplot(data=df, x="Length3", y="Weight", hue="Species")
plt.title("Fish Length vs Weight")
plt.show()
```

Fish weight increases as body length increases, indicating strong biological growth patterns.

## 5. Which Measurements Influence Weight the Most?

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

### Correlation Analysis Insights

The correlation matrix reveals strong positive relationships between fish weight and body measurements, particularly the length variables. Length1, Length2, and Length3 all exhibit correlations above 0.90 with weight, indicating that fish length is a strong predictor of fish mass.

Additionally, the three length variables are highly correlated with each other, suggesting they represent similar dimensional measurements of fish size. Height and width also show positive correlations with weight, though slightly weaker than length measurements.

Overall, these findings suggest that fish body dimensions, particularly length, play a critical role in determining fish weight.

The three length variables (Length1, Length2, and Length3) exhibit extremely high correlations with each other, indicating that they measure very similar aspects of fish body size. Including all three variables in predictive models may introduce multicollinearity.

To simplify the model and avoid redundancy, Length3 can be selected as the representative length feature because it reflects the overall fish length and maintains a strong correlation with fish weight.

## 6. Are There Outliers in Fish Measurements?

```python
sns.boxplot(x=df["Weight"])
plt.title("Weight Outliers")
plt.show()
```

A few fish appear significantly larger than the majority, indicating potential outliers or rare large specimens.

```python
sns.boxplot(data=df, x="Species", y="Weight")
plt.xticks(rotation=45)
plt.title("Weight Distribution by Species")
plt.show()
```

Some species exhibit greater variability in weight, suggesting differences in growth patterns.

## 8. What Is the Relationship Between Height and Weight?

```python
sns.scatterplot(data=df, x="Height", y="Weight")
plt.title("Fish Height vs Weight")
plt.show()
```

Fish height is moderately correlated with weight, suggesting thicker fish tend to weigh more.

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'Weight'
Q1 = df['Weight'].quantile(0.25)
Q3 = df['Weight'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define upper and lower bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_no_outliers = df[(df['Weight'] >= lower_bound) & (df['Weight'] <= upper_bound)]

print(f"Original DataFrame shape: {df.shape}")
print(f"DataFrame shape after outlier removal: {df_no_outliers.shape}")
```

Now that the outliers have been removed, let's visualize the distribution of 'Weight' again to confirm.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df_no_outliers['Weight'])
plt.title("Weight Distribution After Outlier Removal")
plt.show()
```

The `df_no_outliers` DataFrame now contains the data with the identified outliers removed from the 'Weight' column. I will now save this processed DataFrame to a new CSV file for Power BI.

```python
df_no_outliers.to_csv('fish_data_processed.csv', index=False)
print('DataFrame without outliers saved to fish_data_processed.csv')
```
